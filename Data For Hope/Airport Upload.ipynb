```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
airports_df = pd.read_csv("airports.dat")
```

```python
airports_df.head()
```

```python
first_row = {column: column for column in airports_df.columns}
```

```python
airports_df = airports_df.append([first_row])
```

```python
airports_df.tail()
```

```python
airports_df.rename(columns={
    "1": "index",
    "Goroka Airport": "Airport Name",
    "Goroka": "City",
    "Papua New Guinea": "Country",
    "GKA": "IATA Airport Code",
    "AYGA": "ICAO Airport Code",
    "-6.081689834590001": "Latitude",
    "145.391998291": "Longitude",
    "5282": "Altitude",
    "10": "Timezone",
    "U": "DST",
    "Pacific/Port_Moresby": "Tz database",
    "airport": "Type",
    "OurAirports": "Source"
}, inplace=True)
```

```python
airports_df.reset_index(inplace=True)
```

```python
airports_df.drop(columns=["level_0", "index"], inplace=True)
```

```python
airports_df.head()
```

```python
engine = create_engine("sqlite:///trafficking.sqlite" ,echo=False)
conn = engine.connect()
```

```python
airports_df.to_sql("airports", conn, if_exists="replace")
```

```python
pd.read_sql_query("select * from airports;", conn)
```
